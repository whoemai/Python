importando o pandas

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep = ';')
dados
```

---

### Removendo os itens duplicados

```python
list(dados['Tipo'].drop_duplicates())
```

```python
residencial = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Casa de Vila']
residencial
```

---

### Verificando se no dataframe 'dados' existe os itens da lista 'residencial'

```python
selecao = dados['Tipo'].isin(residencial)
selecao
```

```python
dados_residencial = dados[selecao]
dados_residencial.head()
```

---

verificando se houve realmente a alteracao nos dataframe

```python
dados_residencial.shape[0]
```

```python
dados_residencial.index = range(dados_residencial.shape[0])
dados_residencial
```
